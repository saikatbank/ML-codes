```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.array([[1,1.2],[2,1.9],[3,6.4],[4,8.3],[5,10]])
```

```python
print(data)
type(data)
```

```python
x,y=data[:,0],data[:,1]
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
theta_val=[0,0]
iteration=5
alpha=0.001
error=[]
for i in range(iteration):
    sum=0
    print('theta_0')
    for x,y in data:
        sum=sum+(y-(theta_val[0]+theta_val[1]*x))  ## here all xi_0 is 1
    print(sum)
    temp=theta_val[0]+(alpha*sum)
    sum=0
    print('theta_1')
    for x,y in data:
        sum=sum+(y-(theta_val[0]+theta_val[1]*x))*x      # Gradient Descent algorithm code
    print(sum)
    theta_val[1]=theta_val[1]+(alpha*sum)
    theta_val[0]=temp
    err=0
    print(theta_val)
    for x,y in data:
        err=err+(y-(theta_val[0]+theta_val[1]*x))**2
    err=err/5
    error.append(err)
```

```python
ite=[i for i in range (iteration)]
```

```python
plt.figure(figsize=(20,10))
plt.scatter(ite,error)
plt.xlabel('iteration')
plt.ylabel('mean square error')
plt.title('iteration vs error graph')
plt.show()
```
